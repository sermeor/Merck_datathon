```python
df_treatments = pd.read_csv('/content/drive/Shared drives/MSD 23 Drive/Datasets/treatments_2017-2020.csv')
```

```python
##'DAYWAIT'
##'PREG'
##'SRVC33'
##Â GENDER

women = df_treatments[df_treatments['GENDER'] == 2]
pregnant_women = women[(women['PREG'] == 1) & (women['DAYWAIT'] != -9)]
nonpregnant_women = women[(women['PREG'] == 2) & (women['DAYWAIT'] != -9)]
```

```python
yp, xp, _ = plt.hist(pregnant_women['DAYWAIT'], bins=[0, 1, 2, 3, 4, 5]);
yn, xn, _ = plt.hist(nonpregnant_women['DAYWAIT'], bins=[0, 1, 2, 3, 4, 5]);
```

```python
yp = yp/sum(yp)
yn = yn/sum(yn)
```

```python
plt.bar(xp[1:], yp, alpha = 0.5, label = 'Pregnant')
plt.bar(xn[1:], yn, alpha = 0.5, label = 'Non pregnant')
plt.legend(frameon = False)
```
